```python
from matplotlib import pyplot as plt
import matplotlib
import numpy as np
import astropy.units as u
```

```python
data = np.loadtxt("/Users/mahaixia/Documents/GitHub/FOF/VIPERS_W4.txt_clusters_0.2_0.01_spec.dat")
udeg = 0.017453293  #deg to rad

RA = data[:,1]
DEC = data[:,3]
z = data[:,5]
```

```python
plt.scatter(RA, DEC)
```

```python
r = data[:,9]   #Mpc
color = data[:,7]   #number of galaxies
color = color
area = r*r*500000
```

```python
fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(projection='polar')

colors= ["limegreen"] * 10 + ["gold"] * 5 + ["red"] * 2
cmap=matplotlib.colors.ListedColormap(colors)

sc = ax.scatter(RA*udeg, z, c=color, s=area, cmap=cmap, alpha=0.75, label=color)

cax = fig.add_axes([0.95,0.42,0.01,0.15])
fig.colorbar(sc, cax=cax, label="N_gal", ticks=[3,5,7,9])

ax.set_thetamax(np.max(RA))
ax.set_thetamin(np.min(RA))
ax.set_rorigin(-0)
ax.set_theta_zero_location('E', offset=205)
ax.set_rmax(0.8)
ax.set_rmin(0.5)
```
